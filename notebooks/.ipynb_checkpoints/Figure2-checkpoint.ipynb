```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world.to_crs("ESRI:54009") 
dfc = gpd.read_file('../data_revision/cities/all/gdfCities.shp')
```

```python
### load the data
cols=['id','city','climate','year','TE', 'pop', 'TEd', 'TEn']
dftemporal = pd.DataFrame(columns=cols)
years = np.arange(2010,2021)
for idx in range(0, 200):
    dff = pd.read_csv('../data_revision/exposure/dfE_%d.csv'%idx)
    dff = dff.dropna().reset_index(drop=True)
    for year in years:
        string1 = 'pop%d'%year
        string2 = 'day%d'%year
        string3 = 'night%d'%year
        dff1 = dff[dff[string1]>=0].reset_index(drop=True)
        if string1 not in dff.columns: 
            print(string1, idx)
            continue
        if string2 not in dff.columns: 
            print(string2, idx)
            continue
        if string3 not in dff.columns: 
            print(string3, idx)
            continue
        E = 0.5*(np.sum(dff1[string1]*dff1[string2])+np.sum(dff1[string1]*dff1[string3]))
        Ed = np.sum(dff1[string1]*dff1[string2])
        En = np.sum(dff1[string1]*dff1[string3])
        data = [idx,dfc['UC_NM_MN'][idx], dfc['climate'][idx], year, E, dff[string1].sum(), Ed, En]
        dftemporal = dftemporal.append(pd.DataFrame(columns=cols, data=[data]),ignore_index=True)
```

```python
dftemporal['TEP']=dftemporal['TE']/dftemporal['pop']
dftemporal_= dftemporal.groupby(['year'])['TE','TEd','TEn'].sum().reset_index()
```

```python
dftemporal_g = dftemporal.groupby('city')['TE','TEP'].max().reset_index()
dftemporal_g['X']=dfc['geometry'].centroid.x
dftemporal_g['Y']=dfc['geometry'].centroid.y
```
